```python
from PIL import Image
import os 
import cv2
from io import BytesIO
import csv
import numpy as nm
import time as t

class Data:
    def __init__(self, path):
        imageObject  = Image.open(path)
        
        img_gray = cv2.cvtColor(nm.array(imageObject ), cv2.COLOR_BGR2GRAY)
        self.citra=img_gray 
        self.matriksIterasi = [[0 for i in range(256)] for j in range(256)]
        
    def insertoCsv(self,data):
        row = data;
        with open('Dataset.csv', 'a') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(row)
        print("succes")
        csvFile.close()
```

```python
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as nm

class Glcm:
    def __init__(self):
        self.result =[]  
    def forOder(self,a, b):
        # size = 0
        # for i in a[0]:
        #     size=size+1
        for i in range(len(a)):
                for j in range(len(a[i])-1):
                    p=a[i][j]
                    q=a[i][j+1]
                    b[p][q]=b[p][q]+1
        matriksIterasi1=list(map(list,zip(*b)))
        for i in range(len(b)):
                for j in range(len(b[0])):
                    b[i][j]=float(b[i][j])+matriksIterasi1[i][j]
        count=0
        for i in range(len(b)):
            for j in range(len(b[0])):
                count=count+b[i][j]
#         print(b)
        return self.normalisasi(b, count)

    def forOder45(self,a, b):
        # size = 0
        # for i in a[0]:
        #     size=size+1
        for i in range(len(a)-1):
                for j in range(len(a[i])-1):
                    p=a[i][j+1]
                    q=a[i+1][j]
                    b[p][q]=b[p][q]+1
        matriksIterasi1=list(map(list,zip(*b)))
        for i in range(len(b)):
                for j in range(len(b[0])):
                    b[i][j]=float(b[i][j])+matriksIterasi1[i][j]
        count=0
        for i in range(len(b)):
            for j in range(len(b[0])):
                count=count+b[i][j]
        return self.normalisasi(b, count)
    
    def forOder90(self,a, b):
        # size = 0
        # for i in a[0]:
        #     size=size+1
        for i in range(len(a)-1):
                for j in range(len(a[i])):
                    p=a[i][j]
                    q=a[i+1][j]
                    b[p][q]=b[p][q]+1          
        matriksIterasi1=list(map(list,zip(*b)))
        for i in range(len(b)):
                for j in range(len(b[0])):
                    b[i][j]=float(b[i][j])+matriksIterasi1[i][j]
        count=0
        for i in range(len(b)):
            for j in range(len(b[0])):
                count=count+b[i][j]
        return self.normalisasi(b, count)

    def forOder135(self,a, b):
        # size = 0
        # for i in a[0]:
        #     size=size+1
        for i in range(len(a)-1):
                for j in range(len(a[i])-1):
                    p=a[i][j]
                    q=a[i+1][j+1]
                    b[p][q]=b[p][q]+1

        matriksIterasi1=list(map(list,zip(*b)))
        for i in range(len(b)):
                for j in range(len(b[0])):
                    b[i][j]=float(b[i][j])+matriksIterasi1[i][j]
        count=0
        for i in range(len(b)):
            for j in range(len(b[0])):
                count=count+b[i][j]
        return self.normalisasi(b, count)

    def normalisasi(self,data, counter):
#         print("Matriks coocurence")
        for i in range(len(data)):
                for j in range(len(data)):
                     data[i][j]=data[i][j]/counter
        return data

    def asmMethod(self, data):
        sume =0
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                sume += (data[i][j]*data[i][j])
        return sume
    
    
    def contrasMethod(self, data):
        result = 0
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                result += data[i][j]*((i-j)*(i-j))
        return result
    
    def idmMethod(self, data):
        result = 0
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                kiri = 1/(1+((i-j)*(i-j)))
                tot = kiri*data[i][j]
                result += tot
        return result

    # def ent2Method(self, data):
    #     result = 0;
    #     for i in range(len(data)-1):
    #         for j in range(len(data[i])-1):
    #             if(data[i][j]>0.0):
    #                 log = math.log2(data[i][j])
    #             else:
    #                 log = 0
    #             value = (data[i][j]*log)*(-1)
    #             result = result + value
    #     return result
    
    def energy(self,asm):
        return math.sqrt(asm)
    
#     new correlation method
    def getMean1(self, data):
        tambah1=0
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                tambah1+=j*data[i][j]
        # print(tambah1)
        return tambah1
    def getMean2(self, data):
        tambah2=0
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                tambah2=i*data[i][j]
        # print(tambah2)
        return tambah2
    def getStd(self, data, u1, u2):
        pangkat1=0
        pangkat2=0
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                pangkat1+=((i-u1)*(i-u1))*data[i][j]
                pangkat2+=((j-u2)*(j-u2))*data[i][j]
        rho1=math.sqrt(pangkat1)
        rho2=math.sqrt(pangkat2)
        # print(rho1*rho2)
        return rho1*rho2
                
    def correlation(self, data):
        hasil=0
        u1 = self.getMean1(data)
        u2 = self.getMean2(data)
        bawah = self.getStd(data, u1, u2)
        for i in range(len(data)-1):
            for j in range(len(data[i])-1):
                tmp = ((i-u1)*(j-u2))/bawah
                akhir = data[i][j]*tmp
                hasil+=akhir
        # print(hasil)
        return (hasil)
        
```

```python
def countfeture(data, glcm):
    asm = glcm.asmMethod(data)
    energy = glcm.energy(asm)
    feture.append(energy)
    idm = glcm.idmMethod(data)
    feture.append(idm)
    cont = glcm.contrasMethod(data)
    feture.append(cont)
    corel = glcm.correlation(data)
    feture.append(corel)
    
```
